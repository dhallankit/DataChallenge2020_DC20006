```python
def season(c):
    if (c >= 3) and (c <= 5):
        return 'Spring'
    elif (c >= 6) and (c <= 8):
        return 'Summer'
    elif (c >= 9) and (c <= 11):
        return 'Fall'
    else:
        return 'Winter'
    

#new column season    
ocean['Season'] = ocean['Month'].apply(season)
```

```python
ocean['Season']
```

```python
#split the state column
ocean['State'] = ocean["State"].str.split(",", n = 1, expand = True) 
```

```python
NorthEast = ['Maine', 'Delaware', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island',
             'Maryland', 'Connecticut','New York', 'New Jersey', 'Pennsylvania']
SouthEast = ['Arkansas', 'Mississippi', 'Louisiana', 'Alabama', 'Georgia', 'Tennessee', 
             'North Carolina', 'South Carolina','West Virginia', 'Virginia', 'Kentucky', 'Florida']
SouthWest = ['Arizona', 'New Mexico', 'Texas', 'Okalahoma']
West = ['California', ' Nevada', 'Oregon', 'Utah', 'Colorado', 'Wyoming', 'Idaho', 'Montana', 'Washington', 'Alaska']
MidWest = ['North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Missouri', 'Iowa', 
           'Minnesota', 'Illinois', 'Indiana', 'Michigan', ' Ohio', 'Wisconsin']

def region(r):
    if r in NorthEast:
        return 'NorthEast'
    elif r in SouthEast:
        return 'SouthEast'
    elif r in SouthWest:
        return 'SouthWest'
    else:
        return 'West'
    

#new column region    
ocean['Region'] = ocean['State'].apply(region)
```
