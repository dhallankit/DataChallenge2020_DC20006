```python
ocean['State'] = ocean['State'].str.replace(', USA', '')
ocean.head()
ocean.shape
```

```python
west = ["Alaska", "Arizona", "California", "Hawaii", "Idaho","Montana","Nevada","Oregon","Utah","Washington", "Wisconsin", "Wyoming"]
midwest = ["North Dakota","Illinois", "Indiana", "Iowa", "Kansas","Michigan","Minnesota","Nebraska","New Mexico","Ohio","South Dakota"]
northeast = ["Connecticut","Maine","Massachusetts","Missouri","New Hampshire","New Jersey","New York", "Pennsylvania", "Rhode Island","Vermont"]
south = ["South Carolina","Alabama","Arkansas", "Delaware","District of Columbia","Florida", "Georgia", "Kentucky", "Louisiana","Maryland","Mississippi","Oklahoma","South Carolina","Tennessee", "Texas","West Virginia","Virginia"]
```

```python
def region(r):
    if r in west: 
        return 'West'
    elif r in midwest:
        return 'Mid-West'
    elif r in northeast:
        return 'North-East'
    elif r in south:
        return 'South'
        
ocean['Region'] = ocean['State'].apply(region)
```

```python
ocean[['Lat', 'Long']] = ocean.GPS.str.split(",", expand = True)
ocean.head()
```

```python
ocean['Region'].value_counts()
```

```python
def season(c):
    if (c >= 3) and (c <= 5):
        return 'Spring'
    elif (c >= 6) and (c <= 8):
        return 'Summer'
    elif (c >= 9) and (c <= 11):
        return 'Fall'
    else:
        return 'Winter'
    

#new column season    
ocean['Season'] = ocean['Month'].apply(season)
ocean.head()
```
